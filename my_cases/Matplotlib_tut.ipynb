```python
import numpy as np
```

```python
x=np.linspace(0,10,20)
```

```python
x
```

```python
y2=x**2
```

```python
y1=x**3
```

```python
plt.plot(x,y1,"bo-",linewidth=2,markersize=10)
```

```python
plt.plot(x,y2,"rs-",linewidth=2,markersize=10)
```

```python
x=np.linspace(0,10,20)
y1=x**3
y2=x**2.6
plt.plot(x,y1,"bo-",linewidth=2,markersize=10,label="First")
plt.plot(x,y2,"rs-",linewidth=2,markersize=10,label="Second")
plt.axis([0,10,0,1000])
plt.xlabel("$distance$")
plt.ylabel("$velocity$");
plt.legend(loc="upper left");
```

```python
#log log plot
x=np.logspace(-1,1,20)
y1=x**3
y2=x**2.6
plt.loglog(x,y1,"bo-",linewidth=2,markersize=10,label="First")
plt.loglog(x,y2,"rs-",linewidth=2,markersize=10,label="Second")
plt.axis([0.1,10,0,1000])
plt.xlabel("$distance$")
plt.ylabel("$velocity$");
plt.legend(loc="upper left");
```
