```python
import numpy as np
import matplotlib.pyplot as plt
delta_X=np.random.normal(0,1,(2,5))
```

```python
delta_X
```

```python
plt.plot(delta_X[0],delta_X[1],"ro")
```

```python
X=np.cumsum(delta_X, axis=1)
```

```python
X
```

```python
plt.plot(X[0],X[1],"ro-")
```

```python
X_0=np.array([[0],[0]])
```

```python
delta_X=np.random.normal(0,1,(2,5))
```

```python
X_con=np.concatenate((X_0,np.cumsum(delta_X, axis=1)),axis=1)
```

```python
plt.plot(X_con[0],X_con[1],"go-")
```

```python
plt.plot(X_con[0],X_con[1],"go-")
plt.savefig("rw.pdf")
```

```python
X_0=np.array([[0],[0]])
delta_X=np.random.normal(0,1,(2,100))
X_con=np.concatenate((X_0,np.cumsum(delta_X, axis=1)),axis=1)
plt.plot(X_con[0],X_con[1],"go-")
plt.savefig("rw1.pdf")
```
