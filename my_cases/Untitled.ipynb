```python
def moving_window_average(x, n_neighbors=1):
    n = len(x)
    p=[]
    sum=0
    width = n_neighbors*2 + 1
    x = [x[0]]*n_neighbors + x + [x[-1]]*n_neighbors
    for i in range(n):
        sum=0
        for j in range(width):
            sum=sum+x[i+j]
        p.append(sum/width)
    return(p)
```
